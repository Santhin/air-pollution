```python
import pandas as pd
#Pokazuje wszystkie kolumny bez tzw. trimma
pd.set_option('display.max_columns', None)


import pickle5 as pickle
```

```python
#Wczytujemy słownik StacjePomiarowe - sql
#Wczytujemy słownik stacji pomiarowych z api gioś  - http://api.gios.gov.pl/pjp-api/rest/station/findAll


StacjePomiarowe_dict=pd.read_excel('./dictionaries/StacjePomiarowe.xlsx') 
api_stations=pd.read_json("./dictionaries/stacje_pom_api.json", orient='records')
with open("./dictionaries/SensoryPomiarowe.pkl", "rb") as fh:
    SensoryPomiarowe_dict= pickle.load(fh)
```

```python
#Słownik mający na celu naprawić błąd w nazewnictwie stacji (po uprzednim sprawdzeniu danego parametru/id stacji etc)
ddict = {'Augustów - mobilne ':'Augustów - mobilne','Suwałki, ul. Pułaskiego':'Suwałki, ul. Pułaskiego 26'}
print(ddict)
StacjePomiarowe_dict.replace({"NazwaStacji":ddict}, inplace=True)
```

```python
#Stosujemy left joina, żeby przypisać stacje pomiarowe z sql'a do listy stacji z api
df4 = pd.merge(StacjePomiarowe_dict, api_stations, left_on=['NazwaStacji'], right_on=['stationName'], how='left')
print(len(df4[df4['stationName'].isnull()]))
```

```python
#Tutaj możemy zobaczyć, że brakuje nam aż 38 stacji 
df4[df4['stationName'].isnull()]
```

```python
# musimy zrobić ponownego joina z plikem metadane STACJE - https://powietrze.gios.gov.pl/pjp/archives

metadane=pd.read_excel("./dictionaries/Metadane - stacje i stanowiska pomiarowe.xlsx",sheet_name='STACJE')
df5 = pd.merge(df4, metadane, left_on=['NazwaStacji'], right_on=['Nazwa stacji'], how='left')
#Tutaj widzimy, że tylko 12 stacji, które znajdują się w słowniku StacjePomiarowe SQL są nie poprawne (sprawdzone zostały metade / kody stacji pomiarowych jak i baza w celu sprawdzenia czy dane stacje posiadają pomiary 
# w naszym przypadku owe stacje są kompletnie puste więc możemy je pominąć)
df5[(df5['stationName'].isnull()) &  (df5['Nazwa stacji'].isnull())]
```
