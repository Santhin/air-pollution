```python
def filtr_files():
    def multiple_file_types(*patterns):
        return it.chain.from_iterable(glob.iglob(pattern) for pattern in patterns)
    file_list=[]
    for filename in multiple_file_types(*pattern):
        file_list.append(filename)
    return file_list
```
